```python
selected_columns = ['name', 'price', 'keywords']
wed_df = df[selected_columns]
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

# NaN 값을 빈 문자열로 대체
wed_df['keywords'] = wed_df['keywords'].fillna('')

#CountVectorizer를 적용하기 위해 공백문자로 word단위가 구분되는 문자열로 변환.
count_vect = CountVectorizer(min_df=1, ngram_range=(1,2))
keywords_mat = count_vect.fit_transform(wed_df['keywords'])
```

```python
print(wed_df)
print(wed_df.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

keywords_sim = cosine_similarity(keywords_mat, keywords_mat)
print(keywords_sim.shape)
print(keywords_sim[:1])
```

```python
keywords_sim_sorted_ind = keywords_sim.argsort()[:,::-1]
print(keywords_sim_sorted_ind[:4])
```

```python
def find_sim_wedding(df, sorted_ind, title_name, top_n=10):
    #인자로 입력된 movies_df DataFrame에서 'title'칼럼이 입력된 title_name값인 DataFrame 추출
    title_wed = wed_df[wed_df['name'] == title_name]

    #title_named을 가진 DataFrame의 index객체를 ndarray로 반환하고
    #sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n개의 index추출
    title_index = title_wed.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]

    #추출된 top_n index출력, top_n index는 2차원 데이터임
    #dataframe에서 index로 사용하기 위해서 1차원 array로 변경
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    return wed_df.iloc[similar_indexes]

similar_wedding = find_sim_wedding(wed_df, keywords_sim_sorted_ind, '엘블레스', 10) # 70
```
